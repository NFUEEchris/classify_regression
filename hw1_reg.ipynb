```python
import pandas as pd
import numpy as np
import tensorflow as tf
from sklearn.model_selection import train_test_split
from sklearn import preprocessing
from keras.activations import sigmoid
from keras.optimizers import SGD,Adam
from keras.losses import categorical_crossentropy
from keras.utils import to_categorical
```

```python
data=pd.read_csv('C:\\code\\python_p\\deep learning\\DL 1\\regress1_trn.csv')
y_data=data.iloc[:,10]
x_data=data.iloc[:,:10]
x_data = preprocessing.normalize(x_data)
norm=np.linalg.norm(y_data)
y_data=y_data / norm
X_train, X_test, y_train, y_test = train_test_split(x_data, y_data, test_size=0.2,random_state=40)

# print(X_train)
# print(X_test.shape)
# print(y_train.shape)
# print(y_test.shape)
```

```python
data1=pd.read_csv('C:\\code\\python_p\\deep learning\\DL 1\\regress1_tst.csv',header=None)
X_test=data1.iloc[:,:10]
```

```python
y_train
```

```python
from keras.layers import Dense,Dropout
from keras.models import  Sequential
from sklearn.linear_model import LinearRegression
model = Sequential()

model.add(Dense(32, input_dim=10, activation='relu'))
model.add(Dense(32))
model.add(Dense(32))
model.add(Dense(1))
optimizer = tf.keras.optimizers.RMSprop(0.0099)
model.compile(loss='mean_squared_error',optimizer=optimizer,metrics=['mae'])
```

```python
model.fit(X_train,y_train,epochs=100)
```

```python
pred_y = model.predict(X_test)
```
